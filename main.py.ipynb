```python
import pandas as pd
import numpy as np

train = pd.read_csv('input/train.csv')
test = pd.read_csv('input/test.csv')
```

```python
train.head(10)
```

```python
train['Sex'].replace('female', 0, inplace=True)
train['Sex'].replace('male', 1 ,inplace=True)
```

```python
train.head()
```

```python
train = train[np.isfinite(train['Age'])]
```

```python
train.describe()
```

```python
train
```

```python
y = train['Survived']
```

```python
y.head()
```

```python
features = ['Pclass', 'Sex', 'Age']
```

```python
x = train[features]
x.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
clf = DecisionTreeRegressor(random_state=1)
clf.fit(x, y)
from sklearn.metrics import mean_absolute_error
predicted_outcome = clf.predict(x) #these are the predicted values
print("Train data accuracy: ", mean_absolute_error(y, predicted_outcome))
```
